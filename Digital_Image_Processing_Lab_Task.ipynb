```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread("/content/sample_data/P1.jpg")
img2=cv2.imread("/content/sample_data/P1.jpg")
cv2_imshow(img1)
cv2_imshow(img2)
```

Addition images
=>Noise Reduction

```python
img1=cv2.resize(img1,(300,300))
img2=cv2.resize(img2,(300,300))
img_add=cv2.add(img1,img2)
cv2_imshow(img_add)
```

Subtraction
=>Change detection

```python
img_sub=cv2.subtract(img1,img2)
cv2_imshow(img_sub)
```

Multiplication
=>Masking

```python
img_mul=cv2.multiply(img1,img2)
cv2_imshow(img_mul)
```
